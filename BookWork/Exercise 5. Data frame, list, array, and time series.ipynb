```python
%%R

# 1
# a

tsEwma = function(tsDat, m0=0, delta=0.7){
                  n = length(tsDat)
                  m_vec = c()
           
                  m_i = m0
                  for(i in 1:n){
                      e_i = tsDat[i] - m_i
                      m_vec[i] = m_i + ((1 - delta) * e_i)
                      m_i = m_vec[i]
                      }
                  return(ts(m_vec, start=start(tsDat), frequency=frequency(tsDat)))}

datVec = c(1, 2, 3, 4)
timeseries = ts(datVec, start=c(1960,3), frequency=12)
#print(tsEwma(timeseries))

# b

tsEwma2 = function(tsDat, m0=0, delta=0.7){
                  tsDat = as.vector(tsDat)
                  n = length(tsDat)
                  m_vec = c()
           
                  m_i = m0
                  for(i in 1:n){
                      e_i = tsDat[i] - m_i
                      m_vec[i] = m_i + ((1 - delta) * e_i)
                      m_i = m_vec[i]
                      }
                  return(ts(m_vec, start=start(tsDat), frequency=frequency(tsDat)))}
                  
tmp <- ts(rnorm(4000), start=c(1960,3), frequency=12)
print(system.time(tsEwma2(tmp)))
print(system.time(tsEwma(tmp)))
```

```python
%%R

# 2

# a

myListFn = function(n){
                    x = rnorm(n, mean=0, sd=1)
                    avg = mean(x)
                    if(avg >= 0){
                            y = rexp(n, rate=1/avg)}
                    else{
                            z = rexp(n, rate=-1/avg)
                            y = -z}
                    k = sum(abs(y) > abs(x))
      
                    return(list(xVec=x, yVec=y, count=k))}

# b

# print(lapply( rep(10,4), myListFn ))
# print(sapply( rep(10,4), myListFn ))
```

```python
%%R

# 5

A = matrix(c(0, 1, 2, 4/9, 14/9, 0, 3, 0, 4/3, 4/3), ncol=2)

drawA = function(X){
                lines(X[1:3, 1], X[1:3, 2])
                lines(X[4:5, 1], X[4:5, 2])
                   }

plot(x=c(-10,10), y=c(-10,10))
drawA(A)
```

```python
%%R

# a

shift = function(X, a, b){
                          X[, 1] = X[, 1] + a
                          X[, 2] = X[, 2] + b
                          return(X)}


plot(x=c(-10,10), y=c(-10,10))
shifted = shift(A, -3, -3)
drawA(A)
drawA(shifted)

# b

rotate = function(X, r){
                        r_mat = matrix(c( cos(r), -sin(r), sin(r), cos(r) ), ncol=2)
                        return(X %*% r_mat)}

rotated = rotate(A, pi/2)
drawA(rotated)
```
