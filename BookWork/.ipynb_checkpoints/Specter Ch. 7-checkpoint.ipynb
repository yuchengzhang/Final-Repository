```python
%%R

# some of these functions may need the cluster library

# kmeans(x, centers, iter.max = 10, nstart = 1,
#       algorithm = c("Hartigan-Wong", "Lloyd", "Forgy",
#                     "MacQueen"), trace=FALSE)
# kmeans finds k amount of clusters by constantly rearranging observations in the clusters around the
# k centers until the observations don't move anymore 

# pam(x, k, diss = inherits(x, "dist"), metric = "euclidean",
#    medoids = NULL, stand = FALSE, cluster.only = FALSE,
#    do.swap = TRUE,
#    keep.diss = !diss && !cluster.only && n < 100,
#    keep.data = !diss && !cluster.only,
#    pamonce = FALSE, trace.lev = 0)
# pam function does partitioning around medioids, medioid is a an observation within a cluster for which 
# the sum of the distances between it and all the other members of the cluster are a minimum 

# hclust(d, method = "complete", members = NULL)
# d is the distance matrix
# dist(x, method = "euclidean", diag = FALSE, upper = FALSE, p = 2) finds the distance matrix
# diag, should the diagonal of the distance matrix be printed?, upper, should the upper triangle
# of the distance matrix be printed?, p, power of the Minkowski distance

# agnes(x, diss = inherits(x, "dist"), metric = "euclidean",
#      stand = FALSE, method = "average", par.method,
#      keep.diss = n < 100, keep.data = !diss)
# agnes is like hclust, but uses fewer shortcuts when updating the distance matrix
# For example, when the mean method of calculating the distance between observations and clusters is used, 
# hclust only uses the two observations and/or clusters which were recently merged when updating the distance
# matrix, while agnes calculates those distances as the average of all the distances between
# all the observations in the two clusters
```

```python
%%R

cars = read.delim('http://www.stat.berkeley.edu/classes/s133/data/cars.tab', stringsAsFactors=FALSE)

# read.delim reads tab delimited files

print(head(cars))
```

```python
%%R 

# scale(x, center = TRUE, scale = TRUE), centers and scales the columns of a numeric matrix
# center, either a logical value or a numeric vector of length equal to the number of columns of x
# scale, either a logcial value or a numric vector of length equal to the number of columns of x

# Mean average deviation:
#  1. Find the mean of all values
#  2. Find the absolute value of the difference of each value from that mean (subtract the mean from each value, ignore minus signs)
#  3. Then find the mean of those distances

# Median absolute deviation
#  median of the absolute deviations from the data's median 

car_nums = cars[, -c(1, 2)]
medians = apply(car_nums, 2, median, na.rm=TRUE)
median_avg_deviations = apply(car_nums, 2, mad, na.rm=TRUE)

standardized = scale(car_nums, center=medians, scale=median_avg_deviations)
```

```python
%%R

cars.dist = dist(standardized)

# distance marix is symmetric
# if you want to use distance matrix for anything other then clustering functions you have to use as.matrix
```

```python
%%R

cars.hclust = hclust(cars.dist)

print(cars.hclust)

# by default when a cluster is formed its distance to other objects is computed as the maximum distance
# between any object in the cluster and that other object
```

```python
%%R

plot(cars.hclust, labels=cars$Car, main='Default from hclust')
```

```python
%%R

groups.3 = cutree(cars.hclust, 3)

print(table(groups.3))
```

```python
%%R

counts = sapply(2:6, function(ncl){table(cutree(cars.hclust, ncl))})

print(counts)
```

```python
%%R

print(cars$Car[groups.3 == 1])
```

```python
%%R

groups.4 = cutree(cars.hclust, 4)

print(sapply(unique(groups.4), function(x){cars$Car[groups.4 == x]}))
```

```python
%%R

print(table(groups.3, cars$Country))
```

```python
%%R

# aggregate function performs summaries on many variables simultaneously 
# it Splits the data into subsets, computes summary statistics for each, and returns the result in a convenient form.
# demands that the variables used to divide up the data are passed to it in a list 

print(aggregate(car_nums, list(groups.3), median))
```

```python
%%R

a3 = aggregate(cars[, -c(1, 2)], list(groups.3), median)
print(data.frame(Clusters=a3[,1], Freq=as.vector(table(groups.3)), a3[,-1])) 

# as.vector makes the table into a vector of frequencies
```

```python
%%R

a4 = aggregate(cars[, -c(1,2)], list(groups.4), median)
print(data.frame(Clusters=a4[,1], Freq=as.vector(table(groups.4)), a4[,-1]))
```

```python
%%R

cars.pam = pam(cars.dist, 3)

# can use table to compare results of hclust and pam functions
print(table(groups.3, cars.pam$clustering))
```

```python
%%R

print(cars$Car[groups.3 != cars.pam$clustering])
```

```python
%%R

print(cars$Car[cars.pam$id.med]) 

# id.med contains indexes of the medoids
```

```python
%%R

# silhouette(x, ...)
# x has an x$clustering component
# creates a graph showing how well each observation fits the cluster it is assigned to
# does this by comparing how close an object is to other objects in its own cluster with how close it
# is to objects in other clusters
# values near one mean the observation is well placed in its cluster, values near 0 mean the observation
# could likely belong to some other cluster
# silhouette plot is useful in locating groups in a cluster analysis that may not be doing a good job

plot(cars.pam)
```

```python
%%R

# making a silhouette plot for the 4 group hierarchical cluster

plot(silhouette(groups.4, cars.dist))

# takes in the cut cluster, and a distance matrix used for the clustering 
```

```python
%%R

iris.use = subset(iris, select=-Species)

# subset(x, subset, select)
# x = your data, subset = what you want from the data, select = which columns you want to keep
d = dist(iris.use)
d1 = daisy(iris.use) # another function to calculate distance matrix 

print(sum(abs(d - d1)))
```

```python
%%R

dd = dist(iris.use, method='manhattan')
print(sum(abs(as.matrix(dd) - as.matrix(d1))))
```

```python
%%R

z = agnes(d)
plot(z, which.plots=1)

# 1 is a banner plot, white area on the left represents unclustered data, white lines that stick
# into red show heights at which the clusters were formed, red is the clusters
# if you cut at a height, you figure out amount of clusters by finding out how many red pieces are separated
# ie if you cut at two you get 3 clusters 
# banner plot is an alternative to a dendogram
```

```python
%%R

plot(z, which.plots=2)

# second plot is a dendogram
# dendogram shows the same relationships as the banner plot
```

```python
%%R

# lets see how well the clusters do in grouping the Irises by species

print(table(cutree(z, 3), iris$Species))
```
