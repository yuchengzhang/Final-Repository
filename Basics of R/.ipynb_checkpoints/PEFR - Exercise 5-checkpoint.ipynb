```python
%%R

# 1

# a

tsEwma = function(tsDat, m0=0, delta=0.7){
                  len = length(tsDat)
                  m = c(m0, rep(NA, len))
                  for(i in 2:(len+1)){
                      e = tsDat[i-1] - m[i-1]
                      m[i] = m[i-1] + ((1 - delta)*e)}
                      # m[i] = ((1 - delta) * tsDat[i-1]) + (delta * m[i-1]) 
    
                  return(ts(m[-1], start=start(tsDat), frequency=frequency(tsDat)))}
    
datVec = seq(2, 100000, by=2)
x = ts(datVec, start=c(1960,3), frequency=12)
print(system.time(tsEwma(x)))


# b

tsEwma2 = function(tsDat, m0=0, delta=0.7){
                  len = length(tsDat)
                  m = c(m0, rep(NA, len))
                  tsDat1 = as.numeric(tsDat) # or c(tsDat) or unclass(tsDat)
                  for(i in 2:(len+1)){
                      e = tsDat1[i-1] - m[i-1]
                      m[i] = m[i-1] + ((1 - delta)*e)}
                      # m[i] = ((1 - delta) * tsDat[i-1]) + (delta * m[i-1]) 
    
                  return(ts(m[-1], start=start(tsDat), frequency=frequency(tsDat)))}

datVec = seq(2, 100000, by=2)
x = ts(datVec, start=c(1960,3), frequency=12)
print(system.time(tsEwma2(x)))
```

```python
%%R

# 2

# a

myListFn = function(n){
                    x = rnorm(n, mean=0, sd=1)
                    avg_x = mean(x)
                    y = rexp(n, rate=(1/avg_x))
    
                    if(avg_x >= 0){
                        y = y}
                    else{
                        z = rexp(n, rate=(-1/avg_x))
                        y = -z}
      
                    k = sum(abs(x) > abs(y))
    
                    ret = list(x, y, k)
                    names(ret) = c('xVec', 'yVec', 'count')
                    return(ret)}

# b 

# print(lapply( rep(10, 4), myListFn))
# print(sapply( rep(10, 4), myListFn)) # Note that sapply is effectively lapply followed by simplify2array.
                                     # If myListFn has no arguments, then similar results can be obtained with

myList = lapply( rep(10, 1000), myListFn)

# c
yVec_list = rep(NA, 1000)

for(i in 1:1000){
    yVec_list[i] = myList[[i]]['yVec']
    }
 
yVec_list = list(yVec_list)
# print(yVec_list)

yVec_extractor = function(i){
                          return(i[['yVec']])}
            
# print(lapply(myList, FUN=yVec_extractor))
```

```python
%%R

# 5
A = matrix(c(0, 1, 2, 4/9, 14/9, 0, 3, 0, 4/3, 4/3), nrow=5, ncol=2)

drawA = function(X)
{
lines(X[1:3,1], X[1:3,2])
lines(X[4:5,1], X[4:5,2])
}

# a

shift = function(X, a, b){
                 len = length(X[, 1])
                 aVec = rep(a, len)
                 bVec = rep(b, len)
                 shiftMat = matrix(c(aVec, bVec), nrow=len, ncol=2)
                 return(X + shiftMat)}
                         
# b 

rotate = function(X, r){
                  R = matrix(c(cos(r), -sin(r), sin(r), cos(r)), nrow=2, ncol=2)
                  return(X %*% R)}

A = cbind(c(0,1,2,4/9,14/9), c(0,3,0,4/3,4/3))

# c

big_arr = array(dim=c(5, 2, 25))
big_arr[, , 1] = A
for(i in 2:25){
    big_arr[, ,i] = rotate(A, ((2*pi)/24) * i)}

#    print(vapply(1:25, FUN=function(x){
# or                                rotate(A, (pi/24) * (x-1))}, FUN.VALUE=matrix(0, nrow=5, ncol=2)))

# print(big_arr)

# d

scale = function(X, a, b){
                 scaleMat = matrix(c(a, 0, 0, b), nrow=2, ncol=2)
                 return(X %*% scaleMat)}

# e

arArandom = array(dim=c(5, 2, 25))
arArandom[, , 1] = A
for(i in 2:25){
    scale1 = scale(arArandom[, , i-1], sample(10), sample(10))
    rotate1 = rotate(scale1, sample(10))
    shift1 = shift(rotate1, sample(10), sample(10))
    arArandom[, , i] = shift1}

# print(arArandom)
```
